# Financial Risk Analysis Notebook

This notebook analyzes the risk metrics calculated from the financial modeling pipeline.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
plt.style.use('seaborn-v0_8')
sns.set_palette('husl')
```

```python
# Load the processed data
df = pd.read_csv('../dashboard/powerbi_data.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(['Ticker', 'Date'])

print(f"Data shape: {df.shape}")
print(f"Columns: {list(df.columns)}")
print(f"Tickers: {df['Ticker'].unique()}")
```

```python
# Exploratory Data Analysis
print("Data Overview:")
print(df.head())

print("\nRisk Metrics Summary:")
risk_summary = df.groupby('Ticker')[['VaR', 'CVaR', 'Sharpe']].mean()
print(risk_summary)

print("\nVolatility Analysis:")
vol_summary = df.groupby('Ticker')['rolling_volatility'].describe()
print(vol_summary)
```

```python
# VaR Comparison Across Stocks
plt.figure(figsize=(12, 6))
for ticker in df['Ticker'].unique():
    ticker_data = df[df['Ticker'] == ticker]
    plt.plot(ticker_data['Date'], ticker_data['VaR'], label=ticker, alpha=0.7)

plt.title('Value at Risk (VaR) Over Time')
plt.xlabel('Date')
plt.ylabel('VaR')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Rolling Volatility Analysis
plt.figure(figsize=(12, 6))
for ticker in df['Ticker'].unique():
    ticker_data = df[df['Ticker'] == ticker]
    plt.plot(ticker_data['Date'], ticker_data['rolling_volatility'], label=ticker, alpha=0.7)

plt.title('Rolling Volatility (21-day) Over Time')
plt.xlabel('Date')
plt.ylabel('Volatility')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Sharpe Ratio Comparison
plt.figure(figsize=(10, 6))
sharpe_avg = df.groupby('Ticker')['Sharpe'].mean().sort_values(ascending=False)
sharpe_avg.plot(kind='bar', color='skyblue')
plt.title('Average Sharpe Ratio by Stock')
plt.xlabel('Stock Ticker')
plt.ylabel('Sharpe Ratio')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Returns Correlation Matrix
returns_pivot = df.pivot(index='Date', columns='Ticker', values='returns')
correlation_matrix = returns_pivot.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, 
            square=True, linewidths=0.5)
plt.title('Stock Returns Correlation Matrix')
plt.tight_layout()
plt.show()
```

## Summary

This notebook provides an analysis of the risk metrics for the selected Indian stocks:

- **VaR (Value at Risk)**: Measures potential loss over a specific time period
- **CVaR (Conditional VaR)**: Expected loss beyond VaR
- **Sharpe Ratio**: Risk-adjusted return measure
- **Rolling Volatility**: 21-day annualized volatility

Key insights:
- Compare risk profiles across different stocks
- Identify periods of high volatility
- Evaluate risk-adjusted performance
- Analyze correlations for portfolio diversification
